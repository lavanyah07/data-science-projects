```python
# CELL 2 — generate synthetic dataset (300 entries)
import random, csv, re
random.seed(42)

starter_50 = [
(1,"shakespeare","O, wherefore art thou so consumed by care?","Oh, why are you so consumed by worry?"),
(2,"shakespeare","Thou hast betrayed the trust I placed in thee.","You have betrayed the trust I put in you."),
(3,"shakespeare","Wouldst thou that I should tarry longer?","Do you want me to stay longer?"),
(4,"shakespeare","By my troth, I saw him at the moonrise.","Truly, I saw him at moonrise."),
(5,"shakespeare","She spake in riddles, yet none could read her mind.","She spoke in riddles, but no one could understand her thoughts."),
(6,"shakespeare","Hath not the morn a softer blush than thou?","Doesn't the morning have a softer glow than you?"),
(7,"shakespeare","I will procure a mask to hide my visage.","I will get a mask to hide my face."),
(8,"shakespeare","Fain would I tell thee secrets of my heart.","I would gladly tell you the secrets of my heart."),
(9,"shakespeare","Count not the years but the deeds thou doest.","Don't count the years, but the actions you take."),
(10,"shakespeare","Anon I shall return with tidings of joy.","Soon I will return with good news."),
(11,"kjv","Blessed are the meek: for they shall inherit the earth.","Blessed are the humble, for they will inherit the earth."),
(12,"kjv","Thou shalt not bear false witness against thy neighbor.","You must not lie about your neighbor."),
(13,"kjv","Forgive us our trespasses, as we forgive those who trespass against us.","Forgive our sins, as we forgive those who sin against us."),
(14,"kjv","The Lord is my shepherd; I shall not want.","The Lord is my shepherd; I will have everything I need."),
(15,"kjv","Behold, I make all things new.","Look, I make all things new."),
(16,"kjv","Seek ye first the kingdom of God and his righteousness.","First seek God's kingdom and his righteousness."),
(17,"kjv","Let not your heart be troubled: ye believe in God, believe also in me.","Do not let your heart be troubled: if you trust God, trust me too."),
(18,"kjv","Cast thy burden upon the Lord, and he shall sustain thee.","Give your burden to the Lord, and he will support you."),
(19,"kjv","For all have sinned and come short of the glory of God.","Everyone has sinned and fallen short of God's glory."),
(20,"kjv","He that dwelleth in the secret place of the most High shall abide under the shadow of the Almighty.","Those who live close to God will find refuge under his protection."),
(21,"prose_early_modern","I have writ unto thee a letter of explanation.","I have written you a letter of explanation."),
(22,"prose_early_modern","Pray, give me leave to speak plainly upon this matter.","Please, allow me to speak plainly about this issue."),
(23,"prose_early_modern","I took my leave and journeyed hence at dawn.","I left and traveled away at dawn."),
(24,"prose_early_modern","He hath a temper that oft provoketh quarrel.","He has a temper that often causes arguments."),
(25,"prose_early_modern","Wherein lies the benefit of such rare counsel?","Where does the benefit of such uncommon advice lie?"),
(26,"prose_early_modern","I desire thee to mend the hearth ere winter cometh.","I want you to repair the fireplace before winter comes."),
(27,"prose_early_modern","The merchant paid in coin and departed in haste.","The merchant paid with coins and hurried away."),
(28,"prose_early_modern","I cannot but commend thy steadfast devotion.","I can only praise your loyal devotion."),
(29,"prose_early_modern","There is small hope of succor from that quarter.","There is little hope of help from that source."),
(30,"prose_early_modern","I took counsel with my friends and found them true.","I spoke with my friends and found them to be loyal."),
(31,"poetic","Verily, the moon doth silver the river's breast.","Truly, the moon silvered the surface of the river."),
(32,"poetic","The rose that once did blush now paleth in the frost.","The rose that once blushed now fades in the frost."),
(33,"poetic","Whence cometh the wind that whispers of old songs?","Where does the wind come from that whispers old songs?"),
(34,"poetic","Song of sorrow, sing thy lament to the empty shore.","O song of sorrow, sing your lament to the empty shore."),
(35,"poetic","Her eyes were stars that outshone the midnight veil.","Her eyes were stars that shone brighter than the night sky."),
(36,"poetic","The time doth tread upon the petals of the day.","Time walks over the petals of the day."),
(37,"poetic","Anon, the twilight folds its gentle hands upon the field.","Soon, twilight gently covers the field."),
(38,"poetic","A thousand sighs lie sleeping 'neath the willow's shade.","A thousand sighs rest beneath the willow's shade."),
(39,"poetic","Mine heart, a cup o'erflowing with secret grief.","My heart is a cup overflowing with hidden sorrow."),
(40,"poetic","She moves as if the breeze were but her shadow.","She moves as if the breeze were only her shadow."),
(41,"mixed","Thou shalt keep safe the covenant that thou didst swear.","You must keep the promise you once swore."),
(42,"mixed","The court did convene, and matters were set forth in council.","The court met, and the issues were presented at the council."),
(43,"mixed","Hast thou no leisure to attend upon this task?","Do you have no time to work on this task?"),
(44,"mixed","Whereupon he took leave with a solemn bow.","Then he took his leave with a solemn bow."),
(45,"mixed","I compel thee to answer by right and law.","I order you to answer according to the law."),
(46,"mixed","The tapestry was sewn with threads finer than the morning mist.","The tapestry was sewn with threads finer than morning mist."),
(47,"mixed","Lo, the bell doth toll for those that pass away.","Look, the bell rings for those who pass away."),
(48,"mixed","Thou hast cause to rejoice this very day.","You have reason to celebrate today."),
(49,"mixed","Let it be recorded in the scrolls of our town.","Record it in the town records."),
(50,"mixed","I pray thee mark well the lessons of this hour.","Please pay attention to the lessons of this moment.")
]

domains = {
    "shakespeare": [
        "Why are you so anxious today?",
        "You have betrayed the trust I placed in you.",
        "Do you want me to stay longer?",
        "Truly, I saw him at dawn.",
        "She spoke softly but hid her thoughts."
    ],
    "kjv": [
        "Blessed are the humble, for they will inherit the earth.",
        "You must not lie about your neighbor.",
        "Forgive our sins, as we forgive others.",
        "Give your burden to the Lord, and he will support you.",
        "Those who live close to God will find refuge under his protection."
    ],
    "prose_early_modern": [
        "I have written you a letter of explanation.",
        "Please allow me to speak plainly about this issue.",
        "I left at dawn and began my journey.",
        "He has a temper that often causes fights.",
        "There is little hope of help from that source."
    ],
    "poetic": [
        "Truly, the moon silvered the surface of the river.",
        "The rose that once blushed now fades in the frost.",
        "Where does the wind come from that whispers old songs?",
        "A thousand sighs rest beneath the willow's shade.",
        "Her eyes were stars that shone brighter than the night sky."
    ],
    "mixed": [
        "You must keep the promise you once swore.",
        "The court met and the issues were presented at the council.",
        "Do you have time to work on this task?",
        "Record it in the town records.",
        "Please pay attention to the lessons of this moment."
    ]
}

extra_subjects = ["You","I","He","She","We","They","The man","The woman","The child","The farmer"]
extra_verbs = ["love","hate","find","know","want","need","remember","forget","bring","seek","keep","leave","return","speak","ask","tell"]
extra_objects = ["the book","the way","the truth","the red apple","the old house","the sea","the letter","the song","the market","the field","the child","the promise"]

def make_modern_lines(n):
    lines = []
    domain_samples = []
    for d, examples in domains.items():
        domain_samples.extend([(d, ex) for ex in examples])
    while len(lines) < n:
        if random.random() < 0.4 and domain_samples:
            d, ex = random.choice(domain_samples)
            if random.random() < 0.5:
                ex2 = ex.replace("you are", "you're").replace("you have", "you've")
            else:
                ex2 = ex
            lines.append((d, ex2))
        else:
            d = random.choice(list(domains.keys()))
            s = random.choice(extra_subjects)
            v = random.choice(extra_verbs)
            o = random.choice(extra_objects)
            punct = random.choice([".", "?", "."])
            sentence = f"{s} {v} {o}{punct}"
            lines.append((d, sentence))
    return lines

def modern_to_archaic(s):
    s = s.strip()
    repl = {
        r"\bYou\b": "Thou",
        r"\byou\b": "thou",
        r"\byour\b": "thy",
        r"\bYou've\b": "Thou hast",
        r"\byou've\b": "thou hast",
        r"\bare\b": "art",
        r"\bdo\b": "dost",
        r"\bdid\b": "didst",
        r"\bhave\b": "hast",
        r"\bhas\b": "hath",
        r"\bwill\b": "wilt",
        r"\bwould\b": "wouldst",
        r"\bcan\b": "canst",
        r"\bcannot\b": "canst not",
        r"\bnot\b": "naught",
        r"\band\b": "an",
        r"\bfor\b": "forsooth",
        r"\bplease\b": "prithee",
        r"\bThe\b": "Ye",
        r"\bthe\b": "ye",
    }
    for pat, rep in repl.items():
        s = re.sub(pat, rep, s)
    s = re.sub(r"\b(\w{4,})s\b", r"\1eth", s)
    if random.random() < 0.18:
        s = "Verily, " + s
    if random.random() < 0.1:
        s = "Prithee, " + s[0].lower() + s[1:]
    s = re.sub(r"\s+", " ", s)
    s = s.replace(" .", ".").replace(" ,", ",")
    return s

entries = []
for idx, d, s, t in starter_50:
    entries.append({"id": idx, "domain": d, "src_old": s, "tgt_modern": t})

next_id = 51
needed = 300 - len(entries)
modern_pairs = make_modern_lines(needed + 20)
for d, modern in modern_pairs[:needed]:
    archaic = modern_to_archaic(modern)
    entries.append({"id": next_id, "domain": d, "src_old": archaic, "tgt_modern": modern})
    next_id += 1

assert len(entries) == 300

out_path = "multi_domain_300.csv"
with open(out_path, "w", newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=["id","domain","src_old","tgt_modern"])
    writer.writeheader()
    for e in entries:
        writer.writerow(e)

print("Saved dataset to", out_path)
# print preview
for r in entries[:8]:
    print(r["id"], r["domain"], ":", r["src_old"], "=>", r["tgt_modern"])
```

Model & tokenizer setup; dataset split

```python
# CELL 4 — choose model name (change to "t5-base" if you have GPU & time)
import os
from datasets import Dataset
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from sklearn.model_selection import train_test_split

# Explicitly disable Weights & Biases early
os.environ["WANDB_DISABLED"] = "true"

MODEL_NAME = "t5-small"   # demo: t5-small (fast). Use "t5-base" on GPU for better quality.
OUTPUT_DIR = "modernizer_demo"   # checkpoint directory
MAX_SOURCE_LENGTH = 128
MAX_TARGET_LENGTH = 128

# load CSV
df = pd.read_csv("multi_domain_300.csv").dropna().reset_index(drop=True)
train_df, test_df = train_test_split(df, test_size=0.15, random_state=42, stratify=df["domain"])
train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
test_ds  = Dataset.from_pandas(test_df.reset_index(drop=True))

# tokenizer & model
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME)
print("Loaded model:", MODEL_NAME)
print("Train size:", len(train_ds), "Test size:", len(test_ds))
```

Preprocessing & tokenization

```python
# CELL 5 — preprocessing
def preprocess_function(batch):
    inputs = [f"modernize [{d}]: {s}" for d,s in zip(batch["domain"], batch["src_old"])]
    model_inputs = tokenizer(inputs, max_length=MAX_SOURCE_LENGTH, truncation=True, padding="max_length")
    labels = tokenizer(batch["tgt_modern"], max_length=MAX_TARGET_LENGTH, truncation=True, padding="max_length")
    label_ids = labels["input_ids"]
    # replace pad token id's in labels by -100 so loss ignores them
    label_ids = [[(l if l != tokenizer.pad_token_id else -100) for l in lab] for lab in label_ids]
    model_inputs["labels"] = label_ids
    return model_inputs

train_tok = train_ds.map(preprocess_function, batched=True, remove_columns=train_ds.column_names)
test_tok  = test_ds.map(preprocess_function, batched=True, remove_columns=test_ds.column_names)

from transformers import DataCollatorForSeq2Seq
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

Training with auto-resume & checkpointing

```python
# CELL 6 — auto-resume training
import os, numpy as np
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
import evaluate

# metrics
bleu = evaluate.load("sacrebleu")
chrf = evaluate.load("chrf")
def postprocess_texts(preds, labels):
    preds = [p.strip() for p in preds]
    labels = [l.strip() for l in labels]
    return preds, labels

def compute_metrics(eval_pred):
    preds_ids, labels_ids = eval_pred
    preds = tokenizer.batch_decode(preds_ids, skip_special_tokens=True)
    labels_ids = np.where(labels_ids != -100, labels_ids, tokenizer.pad_token_id)
    labels = tokenizer.batch_decode(labels_ids, skip_special_tokens=True)
    preds, labels = postprocess_texts(preds, labels)
    bleu_score = bleu.compute(predictions=preds, references=[[l] for l in labels])["score"]
    chrf_score = chrf.compute(predictions=preds, references=labels)["score"]
    return {"bleu": bleu_score, "chrf": chrf_score}

# find last checkpoint (if exists)
last_checkpoint = None
if os.path.isdir(OUTPUT_DIR):
    chks = [d for d in os.listdir(OUTPUT_DIR) if d.startswith("checkpoint-")]
    if chks:
        chks = sorted(chks, key=lambda x: int(x.split("-")[-1]))
        last_checkpoint = os.path.join(OUTPUT_DIR, chks[-1])
        print("Found checkpoint:", last_checkpoint)
    else:
        print("No checkpoint found in", OUTPUT_DIR)
else:
    print("Output dir does not exist; will create", OUTPUT_DIR)

training_args = Seq2SeqTrainingArguments(
    output_dir=OUTPUT_DIR,
    eval_strategy="epoch",
    save_strategy="epoch",       # save after every epoch
    logging_strategy="steps",
    logging_steps=50,
    learning_rate=3e-4,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    weight_decay=0.01,
    save_total_limit=5,
    num_train_epochs=8,         # total epochs you want
    predict_with_generate=True,
    fp16=False,                 # set True if running on GPU with fp16 support
    load_best_model_at_end=False,
    report_to="none"            # Disable wandb reporting
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_tok,
    eval_dataset=test_tok,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# start or resume
if last_checkpoint:
    print("Resuming training from checkpoint:", last_checkpoint)
    trainer.train(resume_from_checkpoint=last_checkpoint)
else:
    print("Starting training from scratch.")
    trainer.train()

# save final model & tokenizer
trainer.save_model(OUTPUT_DIR)
tokenizer.save_pretrained(OUTPUT_DIR)
print("Training complete. Saved to", OUTPUT_DIR)
```

Interactive user input

```python

# CELL 8 — interactive modernization input (notebook-friendly)
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained(OUTPUT_DIR)
model = AutoModelForSeq2SeqLM.from_pretrained(OUTPUT_DIR)
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

print("Interactive modernization. Enter domain (mixed/shakespeare/kjv/prose_early_modern/poetic).")
domain = input("Domain (press Enter for 'mixed'): ").strip() or "mixed"
if domain not in ["mixed","shakespeare","kjv","prose_early_modern","poetic"]:
    print("Unknown domain; using 'mixed'.")
    domain = "mixed"

print("Paste/enter archaic text. End with an empty line (press Enter twice).")
lines = []
while True:
    try:
        line = input()
    except EOFError:
        break
    if line.strip() == "":
        break
    lines.append(line)
text = " ".join(lines).strip()
if not text:
    print("No input provided — exit cell.")
else:
    prompt = f"modernize [{domain}]: {text}"
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, padding=True).to(device)
    out = model.generate(**inputs, max_length=256, num_beams=6, num_return_sequences=3, early_stopping=True, no_repeat_ngram_size=2)
    outs = tokenizer.batch_decode(out, skip_special_tokens=True)
    print("\nModernized candidates:")
    for i,o in enumerate(outs,1):
        print(f"[{i}] {o}")
```
